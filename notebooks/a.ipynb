```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("aggplot")
```

```python
N = 10000
b = np.random.random(N) * 8.0 + 2.0
h = 2 ** b
v = 2 ** np.random.normal(b, 0.5)
plt.xlabel("h")
plt.ylabel("v")
plt.scatter(h, v, alpha=0.2)
```

```python
w = np.random.random(N) * 49 + 1
p = np.random.normal(w, w / 3)
plt.xlabel("w")
plt.ylabel("p")
plt.scatter(w, p, alpha=0.2)
```
